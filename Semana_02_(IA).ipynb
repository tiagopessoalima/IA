# **Perceptron**

Um dos algoritmos mais antigos e importantes em inteligência artificial, busca  imitar o comportamento básico do neurônio biológico. Ele funciona como um modelo matemático simplificado que tenta replicar a capacidade de processamento e aprendizado do neurônio.



image.png

## **Como funciona?**

É um classificador linear binário que mapeia um vetor de entrada  $x = (x_1, x_2, ..., x_n)$  para uma saída binária  $y ∈ {0, 1}$.  

Ele é definido pela seguinte função:

$y = f(w ⋅ x + b)$

Onde:

- $x$: vetor de entrada com n características.
- $w$: vetor de pesos com n valores, onde cada $w_i$ corresponde ao peso da entrada $x_i$.
- $b$: bias, um valor constante que ajusta a função de ativação.
- $w ⋅ x$: produto escalar entre os vetores $w$ e $x$, calculado como $w_1x_1 + w_2x_2 + ... + w_nx_n$.
- $f$: função de ativação. $f(z) = 1$ se $z ≥ 0$, $0$ caso contrário

## **Como é feito o treinamento?**

Aprende ajustando os pesos $w$ e o bias $b$  através de um algoritmo de aprendizado supervisionado. O algoritmo mais comum é o Regra Delta:

1. **Inicialização:** Inicializa os pesos $w$ e o bias $b$ com valores aleatórios.

2. **Iteração:** Para cada exemplo de treinamento $i$:
- Calcula a saída predita: $ŷi = f(w ⋅ x_i + b)$. O Perceptron calcula a soma ponderada das entradas do exemplo $i$ ($x_i$) com os pesos, adiciona o bias e aplica a função de ativação ($f$) para obter a saída predita ($ŷ_i$).
- Calcula o erro: $e_i = y_i - ŷ_i$. O erro é a diferença entre a saída desejada do exemplo $i$ ($y_i$) e a saída predita ($ŷ_i$).
- Atualiza os pesos: $w_j = w_j + α * e_i * x_{ij}$ para cada $j$. Cada peso ($w_j$) é ajustado proporcionalmente ao erro ($e_i$) e à entrada correspondente ($x_{ij}$) do exemplo $i$. A taxa de aprendizado ($α$) controla o tamanho do ajuste.
- Atualiza o bias: $b = b + α * e_i$. O bias também é ajustado proporcionalmente ao erro do exemplo $i$.

3. **Convergência:** Repete o passo 2 até que o erro seja minimizado ou um número máximo de iterações seja atingido.

Onde $α$ é a taxa de aprendizado, um hiperparâmetro que controla a magnitude da atualização dos pesos.



### **Exemplo: Treinamento para a Porta Lógica AND**

#### **Dados de Treinamento:**

Exemplo | Entrada 1 | Entrada 2 | Saída Desejada
--- | --- | --- | ---
1 | 0 | 0 | 0
2 | 0 | 1 | 0
3 | 1 | 0 | 0
4 | 1 | 1 | 1

#### **1. Inicialização:**

Inicializar os pesos ($w_1$ e $w_2$) e o bias ($b$) com valores aleatórios. Vamos usar:

$w_1 = 0.2$

$w_2 = 0.7$

$b = -0.3$

Definir a taxa de aprendizado ($α$) como $0.1$.

#### **2. Iteração**

Vamos iterar sobre cada exemplo $i$ de treinamento e aplicar a Regra Delta:

##### **Exemplo 1:** (0, 0) -> 0

1. **Calcular a saída predita:**

  $ŷ = degrau(w_1 * x_1 + w_2 * x_2 + b)$

  $ŷ = degrau(0.2 * 0 + 0.7 * 0 + (-0.3))$

  $ŷ = degrau(-0.3) $

  $ŷ = 0$

2. **Calcular o erro:**

  $e = y - ŷ$

  $e = 0 - 0$

  $e = 0$

3. **Atualizar os pesos e o bias:** Como o erro é $0$, os pesos e o bias não serão atualizados neste caso.

##### **Exemplo 2:** (0, 1) -> 0

1. **Calcular a saída predita:**

  $ŷ = degrau(w_1 * x_1 + w_2 * x_2 + b)$

  $ŷ = degrau(0.2 * 0 + 0.7 * 1 + (-0.3))$

  $ŷ = degrau(0.4) $

  $ŷ = 1$

2. **Calcular o erro:**

  $e = y - ŷ$

  $e = 0 - 1$

  $e = -1$

3. **Atualizar os pesos e o bias:**

  $w_1 = w_1 + α * e * x_1$ -> $w_1 = 0.2 + 0.1 * (-1) * 0$ -> $w_1 = 0.2$
  
  $w_2 = w_2 + α * e * x_2$ -> $w_2 = 0.7 + 0.1 * (-1) * 1$ ->$ w_2 = 0.6$
  
  $b = b + α * e$ -> $b = -0.3 + 0.1 * (-1)$ -> $b = -0.4$

##### **Exemplo 3:** (1, 0) -> 0

1. **Calcular a saída predita:**

  $ŷ = degrau(w_1 * x_1 + w_2 * x_2 + b)$

  $ŷ = degrau(0.2 * 1 + 0.6 * 0 + (-0.4))$

  $ŷ = degrau(-0.2)$

  $ŷ = 0$

2. **Calcular o erro:**

  $e = y - ŷ$

  $e = 0 - 0$

  $e = 0$

3. **Atualizar os pesos e o bias:** Como o erro é $0$, os pesos e o bias não serão atualizados neste caso.

##### **Exemplo 4:** (1, 1) -> 1

1. **Calcular a saída predita:**

  $ŷ = degrau(w_1 * x_1 + w_2 * x_2 + b)$

  $ŷ = degrau(0.2 * 1 + 0.6 * 1 + (-0.4))$

  $ŷ = degrau(0.4)$

  $ŷ = 1$

2. **Calcular o erro:**

  $e = y - ŷ$

  $e = 1 - 1$

  $e = 0$

3. **Atualizar os pesos e o bias:** Como o erro é $0$, os pesos e o bias não serão atualizados neste caso.

##### **3. Convergência**:

Após a primeira iteração, repetimos o processo para os quatro exemplos de treinamento várias vezes. O processo continua até que o erro seja zero para todos os exemplos ou que um número máximo de iterações seja atingido.

## **Implementação**

```python
import numpy as np

class Perceptron:
    """
    Classe que implementa um Perceptron.
    """

    def __init__(self, taxa_aprendizado=0.1, epocas=100):
        """
        Inicializa o Perceptron.

        Args:
            taxa_aprendizado (float): Taxa de aprendizado do Perceptron.
            epocas (int): Número de épocas de treinamento.
        """
        self.taxa_aprendizado = taxa_aprendizado
        self.epocas = epocas

    def _ativacao(self, valor):
        """
        Função de ativação degrau.

        Args:
            valor (float): Valor a ser aplicado na função de ativação.

        Returns:
            int: 1 se o valor for maior ou igual a 0, 0 caso contrário.
        """
        return 1 if valor >= 0 else 0

    def _soma_ponderada(self, entradas):
        """
        Calcula a soma ponderada das entradas.

        Args:
            entradas (list): Lista de entradas.

        Returns:
            float: Soma ponderada das entradas.
        """
        return np.dot(entradas, self.pesos) + self.bias

    def fit(self, X, y):
        """
        Treina o Perceptron.

        Args:
            X (list): Lista de listas com as entradas de treinamento.
            y (list): Lista com as saídas desejadas para cada entrada.
        """
        # Inicializa os pesos e o bias com valores aleatórios
        self.pesos = np.random.rand(X.shape[1])
        self.bias = np.random.rand(1)

        # Loop de treinamento
        for _ in range(self.epocas):
            for i in range(len(X)):
                # Calcula a saída do Perceptron
                saida = self._ativacao(self._soma_ponderada(X[i]))

                # Calcula o erro
                erro = y[i] - saida

                # Atualiza os pesos e o bias
                self.pesos += self.taxa_aprendizado * erro * X[i]
                self.bias += self.taxa_aprendizado * erro

    def predict(self, entradas):
        """
        Prediz a saída para uma dada entrada.

        Args:
            entradas (list): Lista de entradas.

        Returns:
            int: Saída predita pelo Perceptron.
        """
        return self._ativacao(self._soma_ponderada(entradas))
```

### **Exercício**: Implementação da Verificação de Convergência

No algoritmo do Perceptron, a convergência ocorre quando o modelo classifica corretamente todos os exemplos de treinamento. Isso significa que o erro é zero para todas as amostras. Atualmente, o algoritmo executa um número fixo de épocas (`self.epocas`), mesmo que ele já tenha convergido antes. O objetivo é modificar o método `fit` para parar o treinamento assim que a convergência for atingida.

```python
import numpy as np

class Perceptron:
    """
    Classe que implementa um Perceptron.
    """

    def __init__(self, taxa_aprendizado=0.1, epocas=100):
        """
        Inicializa o Perceptron.

        Args:
            taxa_aprendizado (float): Taxa de aprendizado do Perceptron.
            epocas (int): Número de épocas de treinamento.
        """
        self.taxa_aprendizado = taxa_aprendizado
        self.epocas = epocas

    def _ativacao(self, valor):
        """
        Função de ativação degrau.

        Args:
            valor (float): Valor a ser aplicado na função de ativação.

        Returns:
            int: 1 se o valor for maior ou igual a 0, 0 caso contrário.
        """
        return 1 if valor >= 0 else 0

    def _soma_ponderada(self, entradas):
        """
        Calcula a soma ponderada das entradas.

        Args:
            entradas (list): Lista de entradas.

        Returns:
            float: Soma ponderada das entradas.
        """
        return np.dot(entradas, self.pesos) + self.bias

    def fit(self, X, y):
        """
        Treina o Perceptron.

        Args:
            X (list): Lista de listas com as entradas de treinamento.
            y (list): Lista com as saídas desejadas para cada entrada.
        """
        # Inicializa os pesos e o bias com valores aleatórios
        self.pesos = np.random.rand(X.shape[1])
        self.bias = np.random.rand(1)

        # Loop de treinamento
        for _ in range(self.epocas):
            erros = 0 # Contador de erros na época
            for i in range(len(X)):
                # Calcula a saída do Perceptron
                saida = self._ativacao(self._soma_ponderada(X[i]))

                # Calcula o erro
                erro = y[i] - saida

                if erro != 0:
                  # Atualiza apenas se houver erro
                  self.pesos += self.taxa_aprendizado * erro * X[i]
                  self.bias += self.taxa_aprendizado * erro
                  erros += 1

            if erros == 0: # Parada antecipada se não houver erro na época
              break

    def predict(self, entradas):
        """
        Prediz a saída para uma dada entrada.

        Args:
            entradas (list): Lista de entradas.

        Returns:
            int: Saída predita pelo Perceptron.
        """
        return self._ativacao(self._soma_ponderada(entradas))
```

### **Exemplo de Uso**

```python
import matplotlib.pyplot as plt

# Função para treinar o perceptron e plotar a reta de decisão
def treinar_e_plotar(X, y, titulo):
    # Treina o Perceptron
    perceptron = Perceptron()
    perceptron.fit(X, y)

    # Plotar a reta de decisão
    w1, w2 = perceptron.pesos
    b = perceptron.bias

    # Gerar os valores da reta
    x_values = np.linspace(-0.5, 1.5, 100)
    y_values = -(w1 * x_values + b) / w2 if w2 != 0 else None

    # Criar regiões coloridas
    x_min, x_max = -0.5, 1.5
    y_min, y_max = -0.5, 1.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    Z = w1 * xx + w2 * yy + b
    plt.contourf(xx, yy, Z > 0, alpha=0.3, colors=["blue", "green"])

    # Plotar os pontos
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color="blue", label="Classe 0")
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color="green", label="Classe 1")

    # Plotar a reta de decisão
    if y_values is not None:
        plt.plot(x_values, y_values, color="red", label="Reta de Decisão")
    else:
        x_reta = -b / w1
        plt.axvline(x=x_reta, color="red", label="Reta de Decisão")

    # Configurações do gráfico
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(titulo)
    plt.grid(True)
    plt.show()
```

## **Limitação do Perceptron**



O Perceptron, enquanto classificador linear, apresenta uma limitação significativa: incapacidade de aprender funções não lineares. Isso implica que ele só consegue classificar corretamente conjuntos de dados que sejam linearmente separáveis. Um conjunto de dados é considerado linearmente separável quando existe um hiperplano capaz de dividir o espaço de características de modo que todos os pontos de uma classe estejam de um lado do hiperplano e todos os pontos da outra classe estejam do lado oposto. Essa restrição torna o Perceptron inadequado para problemas mais complexos, onde a separação entre classes requer fronteiras de decisão não lineares.

### **Construindo a Porta XOR com Perceptrons: Uma Abordagem Multicamadas**


A porta XOR apresenta um desafio para o Perceptron simples devido à sua não linearidade. No entanto, podemos construir a porta XOR combinando as saídas de múltiplos Perceptrons, cada um implementando funções lineares mais simples, como AND e OR.

image.png

#### **1. Tabela verdade da XOR a partir de AND, OR e NOT**



Exemplo | Entrada A | Entrada B  | NOT A | NOT B | A AND (NOT B) | (NOT A) AND B | (A AND (NOT B)) OR ((NOT A) AND B) (XOR)
--- | --- | --- | --- | --- | --- | --- | ---
1 | 0 | 0 | 1 | 1 | 0 | 0 | 0
2 | 0 | 1 | 1 | 0 | 0 | 1 | 1
3 | 1 | 0 | 0 | 1 | 1 | 0 | 1
4 | 1 | 1 | 0 | 0 | 0 | 0 | 0

#### **2. Implementação em código com a classe Perceptron**


```python
# Dados de treinamento para as portas AND, NOT e OR
X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])

X_or = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_or = np.array([0, 1, 1, 1])

X_not = np.array([[0], [1]])
y_not = np.array([1, 0])
```

```python
# Treinando os Perceptrons para as portas AND, OR e NOT
perceptron_and = Perceptron()
perceptron_and.fit(X_and, y_and)

perceptron_or = Perceptron()
perceptron_or.fit(X_or, y_or)

perceptron_not = Perceptron()
perceptron_not.fit(X_not, y_not)
```

```python
# Função para calcular a saída da porta XOR
def xor(a, b):
    not_a = perceptron_not.predict([a])
    not_b = perceptron_not.predict([b])
    and1 = perceptron_and.predict([a, not_b])
    and2 = perceptron_and.predict([not_a, b])
    return perceptron_or.predict([and1, and2])
```

```python
# Testando a porta XOR
print("XOR(0, 0) =", xor(0, 0))
print("XOR(0, 1) =", xor(0, 1))
print("XOR(1, 0) =", xor(1, 0))
print("XOR(1, 1) =", xor(1, 1))
```

**Observação:** Este exemplo demonstra como construir a porta XOR combinando Perceptrons simples. Uma rede neural multicamadas (MLP) com uma camada oculta contendo as portas AND e NOT, e uma camada de saída com a porta OR, seria capaz de aprender a função XOR diretamente, sem a necessidade de definir as funções AND, OR e NOT separadamente
